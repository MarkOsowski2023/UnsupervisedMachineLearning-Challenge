# UNSUPERVISED MACHINE LEARNING

```python
import pandas as pd
import numpy as np

# preprocessing and split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# dimensionality reduction
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# plotting
import matplotlib.pyplot as plt

# clustering
from sklearn.cluster import KMeans
```

```python
# read csv
df = pd.read_csv("Data/myopia.csv")
df.head()
```

```python
# check for NA
df.isnull().sum()
```

## DATA PREP

- drop target variable and keep features
- train test split
- scale xtrain and xtest

```python
# feature variables and target variable

# features
df_features = df.drop(columns="MYOPIC", axis=1)

# target
target = df['MYOPIC']

# fit Standard scaler
scaler = StandardScaler()
scaler.fit(df_features)

# transform train and test
features_scaled = scaler.transform(df_features)
```

## Dimensionality Reduction

- using PCA from sklearn.decomposition module
- keeping 90% of the explained variance
- get explained variance ratio
- reduce demensions further with t-sne
- plot

```python
# PCA
pca = PCA(n_components=0.90) # preserving 90%

# fit with scaled xtrain data
feature_pca = pca.fit_transform(features_scaled)

# explained variance ratio
print(f'Explained Variance Ratio: {pca.explained_variance_ratio_}')
```

```python
# TSNE

# intialize
tsne = TSNE(learning_rate=100)

# fit transform xtrain_pca
feature_tsne = tsne.fit_transform(feature_pca)
```

### Plot

```python
# add transformed features to df_features
df_features['x']  = feature_tsne[:,0]
df_features['y'] = feature_tsne[:,1]
df_features['class'] = df['MYOPIC']
```

```python
# scatter plot
plt.scatter(x = df_features['x'], y = df_features['y'], c=df_features['class'])
plt.show()
```

The plot does not show any distinct clusters

## Cluster Analysis with K-means

- creating an elbow plot to determine best number of Clusters
- visualizing the clusters

```python
# create dataframe with 2 principal components
df_pca = pd.DataFrame(feature_tsne, columns=["pc_1", "pc_2"])
df_pca.head()
```

```python
# using a for loop to deterime best k value, and calculate the inertia for the range of k values

inertia = []

k = list(range(1, 11))

# loop
for i in k:
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_pca)
    inertia.append(kmeans.inertia_)
    
# ELBOW CURVE

# create dataframe for elbow curve
curve_data = {"k": k, "inertia": inertia}

df_curve = pd.DataFrame(curve_data)

# PLOT
plt.plot(df_curve['k'], df_curve['inertia'])
plt.xticks(list(range(11)))
plt.xlabel("K (number of clusters)")
plt.ylabel("inertia")
plt.show()
```

Best K value is either 4 or 5

### Predicting Clusters

```python
# using k = 5

# initialize
km_model = KMeans(n_clusters=5, random_state=42)

# fit with prinicipal components
km_model.fit(df_pca)

# predict
predicted_clusters = km_model.predict(df_pca)

# adding predicted class to df_pca
df_pca['class'] = km_model.labels_

df_pca.head()
```

```python
# plotting the clusters

plt.scatter(x=df_pca['pc_1'], y=df_pca['pc_2'], c=df_pca['class'])
plt.xlabel("principal component 1")
plt.ylabel("principal component 2")
plt.show()
```

Now we have distinct clusters
